```python
#Step 5: Divide the data into independent and dependent variables
X=pd.DataFrame(data['lstat'])
Y=pd.DataFrame(data['medv'])
print(X)
print(Y)
print(X.shape)
print(Y.shape)
```

```python
# Split the data into train and test sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
#  Shape of the train and test sets
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
xtrain=sc_x.fit_transform(X_train)
xtest=sc_x.fit_transform(X_test)
print(xtrain[0:10,:])
```

```python
# Train the algorithm
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# Retrieve the intercept
print(regressor.intercept_)
```

```python
#  Retrieve the slope
print(regressor.coef_)
```

```python
plt.scatter(X_train,Y_train,color='ORANGE')
```

```python
# predicated value
Y_pred= regressor.predict(X_train)
print(Y_pred)
```

```python
# Evaluate the algorithm
from sklearn import metrics
import numpy as np
print("Mean Absolute Error:",metrics.mean_absolute_error(Y_pred,Y_train))
print("Mean Square Error:", metrics.mean_squared_error(Y_pred,Y_train))
print("Root Mean squared Error:",np.sqrt(metrics.mean_squared_error(Y_pred,Y_train)))
```
